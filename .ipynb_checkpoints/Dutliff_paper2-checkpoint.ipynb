```python
def fileName(file):
    return file.split('_')[0]
```

```python
def truePositiveRate (preds, trues):
    trueIndeces = []
    for (i, val) in enumerate(trues):
        if (val == 1):
            trueIndeces.append(i)

    TP = 0
    for i in trueIndeces:
        if (preds[i] == 1):
            TP += 1
            
    FN = 0
    for i in trueIndeces:
        if (preds[i] == 0):
            FN += 1

    return TP/(TP+FN)


def falsePositiveRate (preds, trues):           
    falseIndeces = []
    for (i, val) in enumerate(trues):
        if(val == 0):
            falseIndeces.append(i)
            
    TN = 0
    for i in falseIndeces:
        if (preds[i] == 0):
            TN += 1
            
    FP = 0
    for i in falseIndeces:
        if (preds[i] == 1):
            FP += 1

    return FP / (FP+TN)
```

```python
def parse(row):
    return list(map(lambda x: float(x), row))

def getData(file_path):
    res = []

    with open(file_path) as csvfile:
        r = csv.reader(csvfile, delimiter=',')

        for row in r:
            res.append(row)

    res = list(map(parse, res[1:]))
    
    for row in res:
        row[-1] = int(row[-1])
 
    return np.array(res)
```

```python
def interpolate(data, numAfter):
    x = np.array(list(map(lambda x: x[-1], data)))

    new_x = np.linspace(x.min(), x.max(), numAfter)

    res = []
	
    for i in range(len(data[0]) - 1) :
        y = list(map(lambda x: x[i], data))

        new_y = sp.interpolate.interp1d(x, y, kind='cubic')(new_x)
        res.append(new_y)
        # np.append(res, [new_y])

    return np.array(res).transpose(1, 0)
```

```python
def combineFile(file_name):
	name = fileName(file_name)

	accData = getData(ACC_PATH + file_name)
	gravData = getData(GRAV_PATH+ file_name)
	gyroData = getData(GYRO_PATH + file_name)
	rotData = getData(ROT_PATH + file_name)

	accData = interpolate(accData, NUM_DATA_POINTS)
	gravData = interpolate(gravData, NUM_DATA_POINTS)
	gyroData = interpolate(gyroData, NUM_DATA_POINTS)
	rotData = interpolate(rotData, NUM_DATA_POINTS)

	accumData = []

	for i in range(len(accData)):
		accumData.append(np.concatenate([accData[i], gyroData[i], gravData[i], rotData[i]]))

	return np.array(accumData)
	
```

```python
def fetchData(names):
    retDict = {}

    for file_name in names:
        name = fileName(file_name)
        data = combineFile(file_name)
        
        try:
            retDict[name].append(data)
        except:
            retDict[name] = [data]
    
    return retDict
            
```

```python
def setupData(dic, user, attacker):
    user_data = np.array(dic[user])
    attacker_data = np.concatenate(list(map(lambda x: dic[x], attacker)))

    user_labels = [1] * len(user_data)
    attacker_labels = [0] * len(attacker_data)

    all_labels = np.concatenate((user_labels, attacker_labels))
    all_data = np.concatenate([user_data, attacker_data])

    return (all_data, all_labels)
```

```python
def lpf2(y):
    Fs = 1000; 
    lp = 50;

    m=7000
    
    
    yf = fft(y);
    #print(yf.shape)
    freq = round((lp/Fs) * (m/5));
    #print(freq)
    yf[:,freq:m - freq] = 0;
    return real(ifft(yf));
```

```python
def compute_min_max_mean_feature(mfccs):
    dmfccs = []
    for m in mfccs:
        me=np.mean(m,axis=0)
        std=np.std(m, axis=0)
        kurt=kurtosis(m, axis=0)
        ske=skew(m, axis=0, bias=True)
        ma=np.max(m,axis=0)
        mi=np.min(m,axis=0)
        dm = vstack((mi, ma , me, std, kurt, ske))
        dmfccs.append(dm.flatten())
    return dmfccs
```

```python
def compute_delta_mfccs(mfccs):
    dmfccs = []
    for m in mfccs:
        tmp = m[1:] - m[0:-1]
        dm = hstack((m[0:-1], tmp))
        dmfccs.append(dm)
    return dmfccs
```

```python
def bow_transform(model, mfccs):
    numwords = model.cluster_centers_.shape[0]
    bows = zeros((len(mfccs), numwords))
    for i in range(len(mfccs)):
        w = model.predict(mfccs[i])
        bw = bincount(w, minlength=numwords)
        bows[i,:] = bw
    return bows
```

```python
# plot an ROC curve using class labels and class scores
def plot_roc(tagnames, Yclasses, Yscores):
    fprall = []
    tprall = []
    aucall = []
    for i in range(len(tagnames)):
        fpr, tpr, thresholds = metrics.roc_curve(Yclasses, Yscores[:,i])
        plt.plot(fpr, tpr, lw=0.5, alpha=0.5)
        auc = metrics.auc(fpr, tpr)
        fprall.append(fpr)
        tprall.append(tpr)
        aucall.append(auc)

    # Then interpolate all ROC curves at this points
    all_fpr = unique(concatenate(fprall))
    mean_tpr = zeros_like(all_fpr)
    for i in range(len(tagnames)):
        mean_tpr += interp(all_fpr, fprall[i], tprall[i])

    # Finally average it and compute AUC
    mean_tpr /= len(tagnames)

    # auc of the average ROC curve
    auc = metrics.auc(all_fpr, mean_tpr)

    # average AUC
    mc_auc = mean(aucall)

    plt.plot(all_fpr, mean_tpr, 'k-', lw=2)
    plt.title('MCAUC={:.4f}, AUC={:.4f}'.format(mc_auc, auc))

    plt.grid(True)
```
