```python
def read_data_2(data_path):
    data = []
    # 1.txt
    file_names = os.listdir(data_path)
    file_names.sort(key=lambda x:int(x[:-4]))
    for file_name in file_names:
        file_path = os.path.join(data_path, file_name)
        signal_data = np.loadtxt(file_path,delimiter=',', skiprows=1)
        data.append(signal_data)
    data = np.array(data).transpose(0, 2, 1)
    d_shape = data.shape
    return data.reshape(d_shape[0], 1, d_shape[1], d_shape[2])
```

```python
def compute_delta_mfccs(mfccs):
    dmfccs = []
    for m in mfccs:
        tmp = m[1:] - m[0:-1]
        dm = hstack((m[0:-1], tmp))
        dmfccs.append(dm)
    return dmfccs
```

```python
X_train=X_train.transpose(0,2,1,3)
X_train=X_train.reshape(116, 7300, 9)
X_train.shape
```

```python
# bag of words (effecively)
all_dmfccs = vstack(X_train)
print(all_dmfccs.shape)
```

```python
km = cluster.KMeans(n_clusters=100, random_state=4487)
km.fit(all_dmfccs[0::10])  # subsample by 10 to make it faster
km.cluster_centers_
```

```python
def bow_transform(model, mfccs):
    numwords = model.cluster_centers_.shape[0]
    bows = zeros((len(mfccs), numwords))
    for i in range(len(mfccs)):
        w = model.predict(mfccs[i])
        bw = bincount(w, minlength=numwords)
        bows[i,:] = bw
    return bows
```

```python
X_test=X_test.transpose(0,2,1,3)
X_test=X_test.reshape(155-116, 7300, 9)
```

```python
# convert data to bow
train_bow = bow_transform(km, X_train)
test_bow  = bow_transform(km, X_test)
```

```python
tf_trans = feature_extraction.text.TfidfTransformer(use_idf=True, norm='l1')
train_Xtf = tf_trans.fit_transform(train_bow)
test_Xtf  = tf_trans.transform(test_bow)
```

```python
lr = linear_model.LogisticRegressionCV(Cs=logspace(-4,4,20), cv=5, class_weight='balanced', solver='liblinear')
lr.fit(train_Xtf, y_train)
```

```python
# predict from the model
predY = lr.predict(test_Xtf)

# calculate accuracy
acc = metrics.accuracy_score(y_test, predY)
print("test accuracy=", acc)
```
